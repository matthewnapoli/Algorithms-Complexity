```python
nums = [99999999, 1, 2, 3, 4, 5, 1, 2, 999999999]


def check_mins(nums, low=0, high=len(nums)-1):
    index_s = (low+high) // 2
    if high - low <= 1: 
        return None
    if(nums[index_s-1] >= nums[index_s] and nums[index_s] <= nums[index_s+1]): 
        return nums[index_s]
    elif(nums[index_s-1] < nums[index_s+1]): 
        return check_mins(nums, low, nums[index_s])
    elif(nums[index_s-1] > nums[index_s+1]): 
        return check_mins(nums, nums[index_s], high)
```

For divide and conquer algorithms we know that the runtime is given by:

 $T(n) = a*T(n/b) + f(n)$

 where: <br>
   $a >=1$ is the number of recursive calls we make <br>
   $n/b$ is the "size" of the cursive call with b>1 <br>
   $f(n)$ is the runtime of a single recursive call <br>

 So in this case we have:

  $T(n) = log(n)*T(n/2) + O(1)$ <br>
  or <br>
  $T(n) = O(log(n))$
